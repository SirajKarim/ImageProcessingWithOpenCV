```python
img = cv2.imread('Output.jpeg')
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```python
cv2.imshow("Thresholded Image",thresh1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
```

```python
cv2.imshow("Thresholded Image",thresh2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret,thresh3 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
```

```python
cv2.imshow("Thresholded Image",thresh3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret,thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)
```

```python
cv2.imshow("Thresholded Image",thresh4)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret,thresh5= cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)
```

```python
cv2.imshow("Thresholded Image",thresh5)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret,thresh5= cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
cv2.imshow("Thresholded Image",thresh5)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]
for i in range(6):
 plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
 plt.title(titles[i])
 plt.xticks([]),plt.yticks([])
plt.show()
```
